```python
def serialize_data(train_loader, test_loader):    
    loader_types = ['train', 'test']
    loaders = [train_loader, test_loader]
    for loader_id, loader_type in enumerate(loader_types):
        first = True
        loader = loaders[loader_id]
        datas = []
        targets = []
        
        for data, target in loader:
            if first:
                first = False
                datas = data.cpu().detach().numpy()
                targets = target.cpu().detach().numpy()
            else:
                datas = np.append(datas, data.cpu().detach().numpy(), axis=0)
                targets = np.append(targets, target.cpu().detach().numpy(), axis=0)

        serialized_data_dir = 'PTEEnet/serialized_data'
        if not os.path.exists(serialized_data_dir):
            os.makedirs(serialized_data_dir)

        torch.save(datas, serialized_data_dir + '/datas_' + loader_type + '.pt')
        torch.save(targets, serialized_data_dir + '/targets_' + loader_type + '.pt')
        
                
```

```python
#@title load data
from PTEEnet.cifar10_noisy import CIFAR10_noisy
from torchvision import datasets, transforms
import torch
import numpy as np
import os

def load_data():
    root = 'data/cifar10'
    kwargs = {}
    """
    trainset = CIFAR10_noisy(root=root, train=True, validate=False, download=True, add_noise=False, noise_levels=[0],\
        transform=transforms.Compose([\
                #transforms.RandomCrop(32, padding=4),
                transforms.RandomHorizontalFlip(),
                transforms.ToTensor(),
                transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))]))#,
                #addNoise(noise_levels)]))

    testset = CIFAR10_noisy(root=root, train=False, validate=False, download=True, add_noise=False, noise_levels=[0],\
        transform=transforms.Compose([\
                transforms.ToTensor(),
                transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))]))
    """

    trainset = datasets.CIFAR10(root=root, train=True, download=True,\
            transform=transforms.Compose([\
                    #transforms.RandomCrop(32, padding=4),
                    transforms.RandomHorizontalFlip(),
                    transforms.ToTensor(),
                    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))]))

    testset = datasets.CIFAR10(root=root, train=False, download=True,\
        transform=transforms.Compose([\
                transforms.ToTensor(),
                transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))]))

    train_loader = torch.utils.data.DataLoader(trainset, batch_size=256,\
                    shuffle=True, **kwargs)
    test_loader = torch.utils.data.DataLoader(testset, batch_size=256,\
                    shuffle=False, **kwargs)

    return train_loader, test_loader
    
```

```python
#@title main

"""Main function of the program.

The function loads the dataset and calls training and validation functions.
"""
def main():
    train_loader, test_loader = load_data()       
    serialize_data(train_loader, test_loader)
    print('Finished')
```

```python
main()
```
